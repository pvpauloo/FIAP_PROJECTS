```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_json('Producao.json')

print(df.count())
```

```python
df['Tipo'].value_counts().plot(kind='bar')
plt.title('Contagem por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Frequência')
plt.show()
```

```python
df['Produto'].value_counts().plot(kind='bar')
plt.title('Contagem por Tipo')
plt.xlabel('Produto')
plt.ylabel('Frequência')
plt.show()
```

```python
df.groupby('Ano')['Valor'].sum().plot(kind='line', marker='o')
plt.title('Soma de Valores por Ano')
plt.xlabel('Ano')
plt.ylabel('Soma')
plt.grid(True)
plt.show()
```
